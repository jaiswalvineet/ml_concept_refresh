## A person caught person B

```python
t = np.linspace(0, 40, 1000) #start, finish and points https://www.geeksforgeeks.org/numpy-linspace-python/
```

```python
# distance by person a 
d_a = 2.5 * t
```

```python
# distance by person b
d_b =3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title(' A person caught person B')
plt.xlabel(' time ' )
plt.ylabel(' distance ') 
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_a,'green')
ax.plot(t, d_b,'red')
plt.axvline(x=30, color='blue', linestyle='--')
_ = plt.axhline(y=75, color='blue', linestyle='--')
```
